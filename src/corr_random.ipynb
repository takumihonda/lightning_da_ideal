```python
import numpy as np


# sample 1
p1 = 1000
c1 = 0.0
sig1 = 1.0
x1 = np.random.normal( c1, sig1, p1 )
y1 = np.random.normal( c1, sig1, p1 )

# sample 2
p2 = 2000
c = 40
sig2 = 10.0
x2 = np.random.normal( c, sig2, p2 ) 
y2 = np.random.normal( c, sig2, p2 )

x = np.concatenate( [x1, x2] )
y = np.concatenate( [y1, y2] )
```

```python
import matplotlib.pyplot as plt
plt.scatter( x, y )
print( np.corrcoef( x, y ) )
```
