```python
shape, scale = 2., 2.
gamma = np.random.gamma(shape, scale, p)

gamma_mean = np.mean( gamma )
gamma_std = np.std( gamma, ddof=1 )

#import matplotlib.pyplot as plt
#count, bins, ignored = plt.hist(gamma, 50, density=True)
```
